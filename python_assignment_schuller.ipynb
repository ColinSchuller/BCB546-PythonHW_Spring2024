```python
## 1 ##
## Dr. X: this gets sequences 
## Please properly document this function in the Jupyter notebook 
## Your descriptions of all functions should contain information about what the function does,
## as well as information about the return types and arguments.
def get_sequences_from_file(fasta_fn):
    sequence_data_dict = {}
    for record in SeqIO.parse(fasta_fn, "fasta"):
        description = record.description.split()
        species_name = description[1] + " " + description[2]
        sequence_data_dict[species_name] = record.seq
    return(sequence_data_dict)
```

```python
#Remove the identifier line in the .fasta file
def split_fasta_to_files(fasta_file_path):
    with open(fasta_file_path, 'r') as file:
        lines = file.readlines()
    
    for i, line in enumerate(lines):
        
        output_file_path = f'line_{i+1}.fasta'
        
        with open(output_file_path, 'w') as outfile:
            outfile.write(line)
        print(f'Created file: {output_file_path}')

split_fasta_to_files('penguins_cyb_filtered.fasta')
```

```python
#This solution is not elegant but it gets the job done. Here I open each of the 12 .txt files for MW
# and open the penguins_mass_cytb.csv to write the string from the .txt into the correct column and row of the .csv
import csv

#MW fasta1
def update_csv_with_text(text_file_path, csv_file_path):
    with open(text_file_path, 'r') as text_file:
        text_to_insert = text_file.readline().strip() 

    with open(csv_file_path, mode='r', newline='') as csvfile:
        rows = list(csv.reader(csvfile))  

    if len(rows) > 0 and len(rows[0]) >= 3:
        rows[1][2] = text_to_insert  
    else:
        print("CSV does not have  rows or columns to accept this request")

    with open(csv_file_path, mode='w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerows(rows)

update_csv_with_text('fasta_1_MW.txt', 'penguins_mass_cytb.csv')
```
